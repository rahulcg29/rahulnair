```python
###linear and polynomial regression
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
year=[10,20,30,40,50]
marks=[50,60,70,80,90]
myworld=np.poly1d(np.polyfit(year,marks,6))
myline=np.linspace(10,50,50)
plt.scatter(year,marks)
plt.plot(myline,myworld(myline))
plt.xlabel("year")
plt.ylabel("marks")
plt.show()
```

```python
###linear regression
import numpy as np
from scipy import stats
year=[2011,2012,2013,2014,2015]
marks=[85,87,90,91,92]
slope,intercept,r,p,std_err= stats.linregress(year,marks)
def my_func(year):
    return slope*year+intercept
mymodel=list(map(my_func,year))
plt.scatter(year,marks)
plt.plot(year,mymodel)
plt.show()
```
